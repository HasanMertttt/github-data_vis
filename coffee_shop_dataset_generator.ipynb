# Coffee Chain Synthetic Sales & Customer Dataset Generator

This notebook generates a realistic synthetic dataset containing approximately 2,000 transaction records for a coffee shop chain. It covers daily sales, product categories, branch locations, customer demographics, payment methods, and satisfaction scores.

### Step 1: Import Required Libraries and Set Seed
Import `pandas` and `numpy` for data manipulation, `datetime` for date handling, and set a random seed for reproducible results.

```python
# Import pandas for DataFrame operations
import pandas as pd

# Import numpy for mathematical and probabilistic calculations
import numpy as np

# Import datetime modules to work with timestamps
from datetime import datetime, timedelta

# Set random seed for reproducibility
np.random.seed(42)
```

### Step 2: Define Parameters and Generate Transaction IDs
Set the target row count (~2000) and generate unique transaction identifiers.

```python
# Define total number of rows (target: 2000 transactions)
n_rows = 2000

# Generate unique transaction IDs (e.g., TXN-10000 to TXN-11999)
transaction_ids = [f"TXN-{10000 + i}" for i in range(n_rows)]
```

### Step 3: Generate Transaction Dates and Weekend Indicators
Generate random timestamps over a 6-month period and sort them chronologically.

```python
# Define date range between Jan 1, 2024 and Jun 30, 2024
start_date = datetime(2024, 1, 1)
end_date = datetime(2024, 6, 30)
total_seconds = int((end_date - start_date).total_seconds())

# Generate random timestamps in seconds and add to start_date
random_seconds = np.random.randint(0, total_seconds, size=n_rows)
transaction_dates = [start_date + timedelta(seconds=int(s)) for s in random_seconds]

# Sort transaction dates in chronological order
transaction_dates.sort()

# Determine if transaction date falls on a weekend (Saturday=5, Sunday=6)
is_weekends = [1 if d.weekday() >= 5 else 0 for d in transaction_dates]
```

### Step 4: Define Branches and Product Catalog
Specify coffee shop branch locations, product categories, individual products, and unit prices.

```python
# List of branches and assigned visit probabilities
branches = ['Downtown', 'Financial District', 'Suburban Mall', 'University Campus', 'Airport Terminal']
branch_probs = [0.30, 0.25, 0.20, 0.15, 0.10]
branch_names = np.random.choice(branches, size=n_rows, p=branch_probs)

# Define product catalog dictionary: category -> list of (product_name, price)
products_catalog = {
    'Hot Coffee': [
        ('Espresso', 3.00),
        ('Americano', 3.50),
        ('Cappuccino', 4.50),
        ('Caffè Latte', 4.75),
        ('Caramel Macchiato', 5.25)
    ],
    'Cold Coffee': [
        ('Iced Americano', 3.75),
        ('Iced Latte', 5.00),
        ('Cold Brew', 4.80),
        ('Frappuccino', 5.50)
    ],
    'Tea & Beverages': [
        ('Green Tea', 3.25),
        ('Earl Grey', 3.25),
        ('Chai Latte', 4.50),
        ('Matcha Latte', 5.00)
    ],
    'Bakery & Pastry': [
        ('Butter Croissant', 3.50),
        ('Chocolate Muffin', 3.80),
        ('Blueberry Scone', 3.75),
        ('Almond Croissant', 4.25)
    ],
    'Sandwiches & Wraps': [
        ('Turkey & Cheese Sandwich', 7.50),
        ('Veggie Wrap', 6.80),
        ('Avocado Toast', 8.25)
    ],
    'Merchandise': [
        ('Coffee Beans 250g', 14.50),
        ('Reusable Travel Mug', 18.00)
    ]
}

# Set category selection probabilities
categories = list(products_catalog.keys())
category_probs = [0.35, 0.25, 0.15, 0.15, 0.07, 0.03]

# Sample products and prices based on chosen categories
product_categories = []
product_names = []
prices = []

for _ in range(n_rows):
    cat = np.random.choice(categories, p=category_probs)
    prod_info = products_catalog[cat][np.random.randint(0, len(products_catalog[cat]))]
    product_categories.append(cat)
    product_names.append(prod_info[0])
    prices.append(prod_info[1])
```

### Step 5: Quantities, Total Amount, and Customer Demographics
Generate order quantities, compute total order amount, and create customer demographics.

```python
# Sample purchase quantity per transaction (1 to 4 units)
quantities = np.random.choice([1, 2, 3, 4], size=n_rows, p=[0.70, 0.20, 0.07, 0.03])

# Calculate total transaction amount (unit price * quantity)
prices = np.array(prices)
quantities = np.array(quantities)
total_amounts = np.round(prices * quantities, 2)

# Generate payment methods
payment_methods = np.random.choice(['Credit Card', 'Mobile Pay', 'Cash', 'Gift Card'], size=n_rows, p=[0.50, 0.35, 0.10, 0.05])

# Generate customer age (18 to 70 years old)
customer_ages = np.random.randint(18, 70, size=n_rows)

# Generate customer gender
customer_genders = np.random.choice(['Female', 'Male', 'Non-Binary', 'Prefer not to say'], size=n_rows, p=[0.48, 0.46, 0.04, 0.02])

# Loyalty program membership (1 = Member, 0 = Non-Member)
loyalty_members = np.random.choice([1, 0], size=n_rows, p=[0.40, 0.60])

# Customer satisfaction score (1 to 5 scale)
satisfaction_scores = np.random.choice([1, 2, 3, 4, 5], size=n_rows, p=[0.03, 0.07, 0.15, 0.45, 0.30])
```

### Step 6: Assemble DataFrame and Export CSV
Construct the final Pandas DataFrame and export it to a CSV file.

```python
# Construct pandas DataFrame
df = pd.DataFrame({
    'transaction_id': transaction_ids,
    'transaction_date': transaction_dates,
    'branch_name': branch_names,
    'product_category': product_categories,
    'product_name': product_names,
    'quantity': quantities,
    'price': prices,
    'total_amount': total_amounts,
    'payment_method': payment_methods,
    'customer_age': customer_ages,
    'customer_gender': customer_genders,
    'loyalty_member': loyalty_members,
    'satisfaction_score': satisfaction_scores,
    'is_weekend': is_weekends
})

# Display first 5 rows of the DataFrame
print("DataFrame Shape:", df.shape)
df.head()
```

### Step 7: Inspect Summary Statistics and Export
Check DataFrame data types, missing values, numerical summary, and export to CSV.

```python
# Display concise summary of DataFrame info
df.info()

# Display descriptive statistics for numerical columns
display(df.describe())

# Export DataFrame to CSV file
output_file = 'coffee_shop_sales.csv'
df.to_csv(output_file, index=False)
print(f"Successfully saved synthetic dataset to '{output_file}'")
```
